## GEOGRAPHY VARIABLES

```python
cps_sample = '/home/brian/Documents/CPS/data/clean/cps2018.ft'
cbsa_list = pd.read_feather(cps_sample, columns=['CBSA'])['CBSA'].unique()
csa_list = pd.read_feather(cps_sample, columns=['CSA'])['CSA'].unique()
co_list = pd.read_feather(cps_sample, columns=['COUNTY'])['COUNTY'].unique()
```

```python
excel_file = 'https://www.nber.org/cbsa-csa-fips-county-crosswalk/List1.xls'
df = pd.read_excel(excel_file, 
                   header=2, 
                   skipfooter=3)
```

#### CBSA

```python
cbsa = df[['CBSA Code', 'CBSA Title']].drop_duplicates()
```

```python
cbsa[cbsa['CBSA Code'].isin(cbsa_list)].set_index('CBSA Code').to_dict()
```

#### CSA

```python
csa = df[['CSA Code', 'CSA Title']].drop_duplicates()
```

```python
csa[csa['CSA Code'].isin(csa_list)].astype({'CSA Code': 'int'}).set_index('CSA Code').to_dict()
```

#### COUNTY

```python
co = df[['County/County Equivalent', 'State Name', 
         'FIPS State Code', 'FIPS County Code']].drop_duplicates()
```

```python
co['code'] = co['FIPS County Code'] * 100 + co['FIPS State Code']
```

```python
co['name'] = co['County/County Equivalent'] + ', ' + co['State Name']
```

```python
co[co['code'].isin(co_list)].loc[:,['name', 'code']].set_index('code').to_dict()
```
