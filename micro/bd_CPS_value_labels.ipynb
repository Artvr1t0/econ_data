Create dictionaries with value labels for OCC, OCCD, IND, INDD



```python
import pandas as pd
import os, re

os.chdir('/home/brian/Documents/CPS/data/')
```

```python
occs = pd.read_html('https://www.bls.gov/cps/cenocc2010.htm')[0].iloc[:-1]
```

```python
(occs[~occs['2010 Census code(s)'].str.contains('–')]
 .astype({'2010 Census code(s)': int})
 .set_index('2010 Census code(s)')['Occupation title']
 .to_dict())
```

```python
inds = pd.read_html('https://www.bls.gov/cps/cenind2012.htm')[0].iloc[:-1]
```

```python
(inds[~inds['2012 Census code(s)'].str.contains('–')]
 .astype({'2012 Census code(s)': int})
 .set_index('2012 Census code(s)')['Industry title']
 .to_dict())
```

```python
dd = open('January_2017_Record_Layout.txt', 'r', encoding='iso-8859-1').read()
```

```python
i = (dd.split('PRDTIND1', 1)[-1]
       .split('ENTRIES', 1)[-1]
       .split('PRDTIND2', 1)[0]
       .replace('\t\t\t\t\t', '')
       .replace('\n\n', ''))

result = re.findall('\n?\s*?\t?\s*?(\d+)\s*?\t(.*?)\t?\s*?\t', i)
dict([(int(key), value) for (key, value) in result])
```
