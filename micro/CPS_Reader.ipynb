```python
import pandas as pd
import numpy as np
import re
import struct
```

```python
# Data dictionary 
dd_file = 'data/January_2017_Record_Layout.txt'
cps_file = 'data/may18pub.dat'
dd = open(dd_file, 'r', encoding='iso-8859-1').read()

# Series of interest 
series = ['PRTAGE', 'PESEX', 'PEMLR', 'PENLFACT', 
          'PWSSWGT', 'HRMONTH', 'HRYEAR4']

# Regular expression finds rows with variable location details
p = re.compile('\n(\w+)\s+(\d+)\s+(.*?)\t+.*?(\d\d*).*?(\d\d+)')

# Keep adjusted results for series of interest
srs = [(i[0], int(i[3])-1, int(i[4])) 
       for i in p.findall(dd) if i[0] in series]

# Identify which bytes to read on each line
unpack_len = 0
unpack_fmt = ''
for s in srs:
    start = s[1]
    end = s[2]
    if start > unpack_len:
        unpack_fmt += str(start - unpack_len) + 'x'
    unpack_fmt += str(end - start) + 's'
    unpack_len = end
    
v = struct.Struct(unpack_fmt).unpack_from
```

```python
column_labels = [s[0] for s in srs]

age_start = [s[1] for s in srs if 'AGE' in s[0]][0]
age_end = [s[2] for s in srs if 'AGE' in s[0]][0]
wgt_start = [s[1] for s in srs if s[0]=='PWSSWGT'][0]
wgt_end = [s[2] for s in srs if s[0]=='PWSSWGT'][0]
```
