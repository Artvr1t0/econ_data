IMF API Documentation is [here](http://datahelp.imf.org/knowledgebase/articles/667681-json-restful-web-service)

```python
import requests
import pandas as pd
```

## Parameters/ Settings

```python
url = 'http://dataservices.imf.org/REST/SDMX_JSON.svc/'
param = [('dataset', 'IFS'),
         ('freq', 'M'),
         ('country', 'GB+DE'),
         ('series', 'PCPI_IX')]
series = '.'.join([param[x][1] for x in [1,2,3]])

key = 'CompactData/{}/{}'.format(param[0][1], series)
```

## Make request and convert to annual percent change

```python
r = requests.get('{}{}'.format(url, key)).json()
df = pd.DataFrame()
for s in r['CompactData']['DataSet']['Series']:
    df[s['@REF_AREA']] = pd.Series([float(x['@OBS_VALUE']) for x in s['Obs']],
                                  index=[x['@TIME_PERIOD'] for x in s['Obs']])

df = df.pct_change(12) * 100
```

## Plot results

```python
%matplotlib inline
df.plot(title='Inflation rate (annual percent change in CPI)',
                figsize=(10, 3))
```
